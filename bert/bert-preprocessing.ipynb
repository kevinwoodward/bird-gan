```python
import pickle
import numpy as np
import os
```

```python
from bert_serving.client import BertClient
bc = BertClient()
```

```python
# Returns a list of vectors
def encode_file(path, bert):
    with open(path, 'r') as f:
        return bc.encode(f.readlines())
```

```python
encoded = encode_file('./text/001.Black_footed_Albatross/Black_Footed_Albatross_0001_796111.txt', bc)
```

```python
encoded.shape
```

```python
embedded_captions = {}
import os
for root, dirs, files in os.walk("./text/", topdown=False):
    for name in files:        
        embedded_captions[name] = encode_file(root+'/'+name, bc)
```

```python
embedded_captions
```

```python
with open('embeddedcaptions.pickle', 'wb') as handle:
    pickle.dump(embedded_captions, handle)
```
