```python
dgen = ImageDataGenerator()
```

```python
for root, dirs, files in os.walk("./CUB_200_2011/CUB_200_2011/images/", topdown=False):
    if root != './CUB_200_2011/CUB_200_2011/images/':
        folder = root.split('/')[-1]
        savedir = './CUB_200_2011/CUB_200_2011/images-cleaned-named/' + folder + '/'
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        generator = dgen.flow_from_directory(
                                            './CUB_200_2011/CUB_200_2011/images/' + folder + '/',
                                            shuffle=False,
                                            classes=folder,
                                            target_size=(64, 64),
                                            batch_size=len(files),
                                            save_to_dir=savedir)
        generator.next()
```

```python
for i in generator:
    idx = (generator.batch_index - 1) * generator.batch_size
    print(generator.filenames[idx : idx + generator.batch_size])
```

```python
def keyfunc(v):
    return int(v.split('_')[-2])
```

```python
newfolders = glob("./CUB_200_2011/CUB_200_2011/images-cleaned-named/*")
newfolders = [x.replace('\\', '/') + '/' for x in newfolders]

oldfolders = glob("./CUB_200_2011/CUB_200_2011/images/*")
oldfolders = [x.replace('\\', '/') + '/' for x in oldfolders]
```

```python
for i in range(len(newfolders)):
    oldimages = glob(oldfolders[i] + '*')
    newimages = glob(newfolders[i] + '*')
    oldimages = [x.replace('\\', '/') for x in oldimages]
    newimages = [x.replace('\\', '/') for x in newimages]
    newimages.sort(key=keyfunc)
    assert(len(oldimages) == len(newimages))
    for j in range(len(oldimages)):
        src = newimages[j]
        dst = newimages[j].rsplit('/', 1)[0] + '/' + oldimages[j].rsplit('/', 1)[-1]
        os.rename(src, dst)
```
